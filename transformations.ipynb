## 3. Transform Data With Pandas

```python
# Add customer column to orders
# Select customer_id, first_name, last_name from customers
customers = customers[['customer_id', 'first_name', 'last_name']]
# Create  customer column
customers['customer'] = customers['first_name'] + ' ' + customers['last_name']
# Drop first_name and last_name columns
customers = customers.drop(['first_name', 'last_name'], axis=1)
# Left Join orders with customers on customer_id
orders = pd.merge(orders, customers, on='customer_id', how='left')
# Drop customer_id column
orders = orders.drop('customer_id', axis=1)
orders
```

```python
# Add staff column to orders
# Select staff_id, first_name, last_name from staffs
staffs = staffs[['staff_id', 'first_name', 'last_name']]
# Create  staff column
staffs['staff'] = staffs['first_name'] + ' ' + staffs['last_name']
# Drop first_name and last_name columns
staffs = staffs.drop(['first_name', 'last_name'], axis=1)
# Left Join orders with staffs on staff_id
orders = pd.merge(orders, staffs, on='staff_id', how='left')
# Drop staff_id column
orders = orders.drop('staff_id', axis=1)    
orders
```

```python
# Add store column to orders
# Select store_id, store_name from stores
stores = stores[['store_id', 'store_name']]
# Left Join orders with stores on store_id
orders = pd.merge(orders, stores, on='store_id', how='left')
# Drop store_id column
orders = orders.drop('store_id', axis=1)
orders
```

```python
# Left Join order_items with orders
order_items = pd.merge(order_items, orders, on='order_id', how='left')
order_items
```

```python
# Add brand column to products
# Select brand_id, brand_name from brands
brands = brands[['brand_id', 'brand_name']]
# Left Join products with brands on brand_id
products = pd.merge(products, brands, on='brand_id', how='left')
# Drop brand_id column
products = products.drop('brand_id', axis=1)
products
```

```python
# Add category column to products
# Select category_id, category_name from categories
categories = categories[['category_id', 'category_name']]
# Left Join products with categories on category_id
products = pd.merge(products, categories, on='category_id', how='left')
# Drop category_id column
products = products.drop('category_id', axis=1)
products
```

```python
# Drop list_price from products
products = products.drop('list_price', axis=1)
products
```

```python
# Left Join order_items with products
order_items = pd.merge(order_items, products, on='product_id', how='left')
order_items
```

```python
# Left Join stocks with products
stocks = pd.merge(stocks, products, on='product_id', how='left')
# Drop product_id column
stocks = stocks.drop('product_id', axis=1)
stocks
```

```python
# Left Join stocks with stores
stocks = pd.merge(stocks, stores, on='store_id', how='left')
# Drop store_id column
stocks = stocks.drop('store_id', axis=1)
stocks
```

```python
# Save order_items to excel
order_items.to_excel('order_items.xlsx', index=False)
# Save stocks to excel
stocks.to_excel('stocks.xlsx', index=False)
```
